```python
import numpy as np 
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('diabetic.csv')
```

```python
df.info()
```

```python
for column in df.columns:
    print (column)
    print(df[column].unique())
```

```python
#repalce glucose column zero with nan
df.Glucose =df.Glucose .replace(0,np.nan)
median=df.Glucose .median()
df.Glucose =df.Glucose .fillna(median)
```

```python
#replace BloodPressure column
df.BloodPressure =df.BloodPressure.replace(0,np.nan)
medain=df.BloodPressure.median()
df.BloodPressure=df.BloodPressure.fillna(medain)  
```

```python
#repalce invalid value in SkinThickness column
df.SkinThickness=df.SkinThickness.replace(0,np.nan)
median=df.SkinThickness.median()
df.SkinThickness=df.SkinThickness.fillna(median)
```

```python
#repalce invalid vlaues from the Insulin col
df.Insulin =df.Insulin.replace(0,np.nan) 
median=df.Insulin.median()
df.Insulin=df.Insulin.fillna(median)
```

```python
#replace invalid values from the BMI
df.BMI=df.BMI.replace(0,np.nan)
medain=df.BMI.median()
df.BMI=df.BMI.fillna(median)
```

```python
#repalce invalid values from DiabetesPedigreeFunction
df.DiabetesPedigreeFunction=df.DiabetesPedigreeFunction.replace(0,np.nan)
median=df.DiabetesPedigreeFunction.median()
df.DiabetesPedigreeFunction=df.DiabetesPedigreeFunction.fillna(median)
```

```python
#save the file
df.to_csv('cleandata.csv',index=False)
```
